```python
EJERCICIO #19
#Leer una lista e identifique la sublista más larga dentro de dicha lista que se encuentra ordenada ascendentemente.
#por ejemplo, en la lista [5, 6, 5, 1, 2, 6, 9, 2, 4, 1] la mayor sublista ordenada es la [1, 2, 6, 9].

matriz  =  [[32, 22, 0, 6, 1],
            [52, 11, 24, 72, 48],
            [43, 39, 53, 10, 29],
            [9, 3, 12, 23, 13]]


m = matriz        # m = matriz; matriz ingresada


# prepara las dimensiones de la matriz de salida mo.
mo = [None]*len(m[0])  #numero de columnas de m = numero de columnas mo
for i in range(0,len(m[0])):
    mo[i] = [None]*len(m) #numerro de filas de m = numero de columnas de mo

    


for i in range(len(m)):
    for j in range(-len(m[0]),0): #1 empieza al contrario para comprobar
        mo[-j-1][i] = m[i][j]     # el giro hacia la izquierda.         
                                  #2 # indice de la fila mo hacia arriba
                                  # indice de la columna m hacia abajo.
        
        

# genera las dos matrices.
print()
print("Matriz principal")
for i in range(len(m)):
    print(m[i])

print()    
print("Matriz girada 90° hacia la izquierda")
for i in range(len(mo)):
    print(mo[i])
```
